### Performing Label Encoding on Numerical Features

```python
from sklearn import preprocessing
```

```python
def MultiLabelEncoder(columnlist,dataframe):
    for i in columnlist:

        le_X=preprocessing.LabelEncoder()
        dataframe[i]=le_X.fit_transform(dataframe[i])
```
